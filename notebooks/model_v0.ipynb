```python
# ARIMA imports and setup (works with existing df)
import subprocess, sys
import matplotlib.pyplot as plt
import pandas as pd

try:
    import statsmodels.api as sm
    from statsmodels.tsa.arima.model import ARIMA
except Exception:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "statsmodels==0.14.2"])
    import statsmodels.api as sm
    from statsmodels.tsa.arima.model import ARIMA

pd.set_option("display.width", 140)

# Column configuration based on provided schema
DATE_COL = 'date'
ID_COL = 'gvkey'
PRICE_COL = 'prc' if 'prc' in df.columns else None
RET_COL = 'stock_ret' if 'stock_ret' in df.columns else None

print({'DATE_COL': DATE_COL, 'ID_COL': ID_COL, 'PRICE_COL': PRICE_COL, 'RET_COL': RET_COL})
```

```python
# Prepare columns based on provided schema

DATE_COL = 'date'  # provided
ID_COL = 'gvkey'   # provided

# Explicit targets per your columns
PRICE_COL = 'prc' if 'prc' in df.columns else None
RET_COLS = [c for c in ['ret_eom', 'stock_ret'] if c in df.columns]

print('Using columns:')
print(f"  DATE_COL={DATE_COL}, ID_COL={ID_COL}")
print(f"  PRICE_COL={PRICE_COL}")
print(f"  RET_COLS={RET_COLS}")
```

```python
# Build monthly series for one company (robust, using date/gvkey/prc/stock_ret/ret_eom)

FORCE_ID = None  # e.g., '1003'
min_points_levels = [36, 24, 12, 6, 2]

work_df = df.copy()
work_df[DATE_COL] = pd.to_datetime(work_df[DATE_COL], errors='coerce')
work_df = work_df.dropna(subset=[DATE_COL, ID_COL])

ids_to_try = [FORCE_ID] if FORCE_ID is not None else work_df[ID_COL].astype(str).unique().tolist()

ret_priority = [c for c in ['stock_ret', 'ret_eom'] if c in work_df.columns]

chosen_id, monthly = None, None

for min_pts in min_points_levels:
    best_id, best_series = None, None
    for cid in ids_to_try[:5000]:
        sub = work_df[work_df[ID_COL].astype(str) == str(cid)]
        # Prefer price if available and non-empty; else try returns in priority order
        if PRICE_COL is not None and sub[PRICE_COL].notna().any():
            ts = sub[[DATE_COL, PRICE_COL]].dropna().rename(columns={PRICE_COL: 'price'})
        else:
            ts = None
            for rcol in ret_priority:
                tmp = sub[[DATE_COL, rcol]].dropna().rename(columns={rcol: 'ret'})
                if not tmp.empty:
                    tmp = tmp.sort_values(DATE_COL)
                    tmp['price'] = 100 * (1 + tmp['ret']).cumprod()
                    ts = tmp[[DATE_COL, 'price']]
                    break
            if ts is None:
                continue
        m = ts.set_index(DATE_COL).resample('ME').last().dropna()
        if len(m) >= min_pts and m['price'].std(skipna=True) > 0:
            if best_series is None or len(m) > len(best_series):
                best_id, best_series = cid, m
    if best_id is not None:
        chosen_id, monthly = best_id, best_series
        break

if chosen_id is None:
    raise ValueError("No suitable series found. Ensure 'prc' or 'stock_ret'/'ret_eom' has data.")

print(f"Chosen {ID_COL}: {chosen_id} with {len(monthly)} monthly points. Std={monthly['price'].std():.6f}")
print(monthly.head(), monthly.tail(), sep='\n\n')
```

```python
# Fit ARIMA with short-series fallback, forecast one month, plot

series = monthly['price'].asfreq('M')

# Choose order depending on length
n = len(series.dropna())
order = (1,1,1) if n >= 24 else (1,1,0) if n >= 12 else (0,1,1) if n >= 6 else (0,1,0)
print({'n_obs': n, 'order': order})

results = ARIMA(series, order=order).fit()
print(results.summary())

forecast_res = results.get_forecast(steps=1)
forecast_mean = forecast_res.predicted_mean
forecast_ci = forecast_res.conf_int(alpha=0.05)

next_date = forecast_mean.index[-1]
print(f"Next month forecast date: {next_date.date()}  value: {forecast_mean.iloc[-1]:.4f}")
print(f"95% CI: [{forecast_ci.iloc[-1,0]:.4f}, {forecast_ci.iloc[-1,1]:.4f}]")

fig, ax = plt.subplots(figsize=(9,4))
series.plot(ax=ax, label='History')
forecast_mean.plot(ax=ax, style='r--', label='Forecast')
ax.fill_between(forecast_ci.index, forecast_ci.iloc[:,0], forecast_ci.iloc[:,1], color='r', alpha=0.2, label='95% CI')
ax.set_title(f"ARIMA{order} forecast for {ID_COL}={chosen_id}")
ax.set_xlabel('Date')
ax.set_ylabel('Price (level)')
ax.legend()
plt.tight_layout()
plt.show()
```
